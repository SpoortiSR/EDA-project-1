```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

def plot_boxplots_matplotlib(df):
    fig, axes = plt.subplots(nrows=len(df.columns), ncols=1, figsize=(10, 25))   
    for i, column in enumerate(df.columns):
        axes[i].boxplot(df[column].dropna(), vert=False)  
        axes[i].set_title(column)
    plt.tight_layout()
    plt.show()

plot_boxplots_matplotlib(cols_numerical)
    
    
```

```python
def plot_density_with_mean(data, column_name, title):
    sns.set_style("darkgrid")
    plt.figure(figsize=(8,6))
    sns.distplot(data[column_name])
    plt.axvline(data[column_name].mean(), color='black', label='Mean')
    plt.title("Pdf distribution of " + title)
    plt.legend(shadow=True, fontsize="larger")
    plt.show()


for column in cols_numerical:
    plot_density_with_mean(df1, column, column)
```

```python
observed=pd.crosstab(df1['Specialization'], df1['Gender'], margins=True)
```

```python
fig, ax = plt.subplots(figsize=(16,6))
sns.countplot(data=df1, x='Specialization', hue='Gender')
ax.set_xticklabels(ax.get_xticklabels(),rotation=90);
```

```python
# the hypothesis is false
```

```python
df1['exp'] = pd.to_datetime(df1["DOL"]).dt.year - pd.to_datetime(df1['DOJ']).dt.year
```

```python
df1.head()
```

```python
new=df1[(df1["Designation"].isin(["programmer analyst","software engineer","hardware engineer","associate engineer"])) & (df1["exp"]==0)]
plt.figure(figsize=(10,5))
sns.barplot(x="Designation",y="Salary",hue="Gender",data=new)
```

```python
from scipy import stats as st
from scipy.stats import chi2_contingency as cst
pv = st.ttest_1samp(new['Salary'],popmean=250000)[1]
#p- value should be >0.05 to claim the null hypothesis
if pv < 0.05:
    print('We reject the null hypothesis and Average salary is not equal to 250000')
else:
    print('We fail to reject null hypothesis and Avergae salary is equal to 250000')
```

```python
# the pypothesis is wrong ,we reject it.
```

```python
sample_columns = pd.crosstab(df1['Gender'],df1['Specialization'],margins=True)
pv = cst(sample_columns)[1]
if pv < 0.05:
    print('We accept  the null hypothesis and Gender impacts specialization')
else:
    print('We  reject null hypothesis and Gender does not impact specialization')
```

```python
# gender does not impact specialization .
```
